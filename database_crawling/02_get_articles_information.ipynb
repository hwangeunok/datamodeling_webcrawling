## **02. 기사**

```python
import requests 
from bs4 import BeautifulSoup
import time
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
import webdriver_manager
from webdriver_manager.chrome import ChromeDriverManager
import pprint
import re
import pandas as pd
import urllib3
urllib3.disable_warnings()
```

```python
news = pd.read_csv("headline_ranking_news.csv")
```

```python
service = ChromeService(executable_path=ChromeDriverManager().install())
options = Options()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36")
driver = webdriver.Chrome(service=service, options=options)

rankingnews_link_list = list(news['rankingnews_link'])
rankingnews_eval = {}

for rankingnews_link in rankingnews_link_list:
    url = rankingnews_link
    driver.get(url)
    eval = []
    try:
        element = driver.find_element(By.CSS_SELECTOR, "#likeItCountViewDiv > ul")  
        # 쏠쏠정보
        useful = element.find_element(By.CSS_SELECTOR, "li.u_likeit_list.useful > a > span.u_likeit_list_count._count").text
        eval.append(int(useful.replace(',','')))
        
        # 흥미진진
        wow = element.find_element(By.CSS_SELECTOR, "li.u_likeit_list.wow > a > span.u_likeit_list_count._count").text 
        eval.append(int(wow.replace(',','')))
        
        # 공감백배
        touched = element.find_element(By.CSS_SELECTOR, "li.u_likeit_list.touched > a > span.u_likeit_list_count._count").text
        eval.append(int(touched.replace(',','')))
        
        # 분석탁월
        analytical = element.find_element(By.CSS_SELECTOR, "li.u_likeit_list.analytical > a > span.u_likeit_list_count._count").text
        eval.append(int(analytical.replace(',','')))
        
        # 후속강추
        recommend = element.find_element(By.CSS_SELECTOR, "li.u_likeit_list.recommend > a > span.u_likeit_list_count._count").text
        eval.append(int(recommend.replace(',','')))

        time.sleep(2)
    except:
        pass
    
    rankingnews_eval[url] = eval
```

**분류와 바이라인**

```python
service = ChromeService(executable_path=ChromeDriverManager().install())
options = Options()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36")
driver = webdriver.Chrome(service=service, options=options)

rankingnews_link_list = list(news['rankingnews_link'])
rankingnews_category = {}

for rankingnews_link in rankingnews_link_list:
    url = rankingnews_link
    driver.get(url)
    
    try:
        element = driver.find_element(By.CSS_SELECTOR, "#contents")  
        
        categorized = [element.text for element in element.find_elements(By.CSS_SELECTOR, "div.media_end_categorize > a > em")]
        
        bylines = [element.text for element in element.find_elements(By.CSS_SELECTOR, "div.byline > p > span")]
        
        time.sleep(2)
    except:
        pass
    
    rankingnews_category[url] = {'category': categorized,
                                 'byline': bylines}
```
