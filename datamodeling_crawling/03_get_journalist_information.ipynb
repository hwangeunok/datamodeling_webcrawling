```python
import requests 
from bs4 import BeautifulSoup
import time
import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.support.wait import WebDriverWait
import webdriver_manager
from webdriver_manager.chrome import ChromeDriverManager
import pprint
import re
import pandas as pd
import urllib3
urllib3.disable_warnings()
```

```python
news = pd.read_csv("headline_ranking_news.csv")
```

```python
service = ChromeService(executable_path=ChromeDriverManager().install())
options = Options()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36")
driver = webdriver.Chrome(service=service, options=options)

rankingnews_link_list = list(news['rankingnews_link'])
rankingnews_journalist = {}

for rankingnews_link in rankingnews_link_list:
    url = rankingnews_link
    driver.get(url)
    try:
        elements = driver.find_elements(By.CSS_SELECTOR, "#JOURNALIST_CARD_LIST > div.media_journalistcard_item._SUBSCRIBE_ELE > div > div.media_journalistcard_intro > div")

        names = []
        subscriber_numbers = []
        cheer_numbers = []

        for i in range(len(elements)):
            
            name = elements[i].text.split(" 기자", 1)[0]
            
            subscriber_number = elements[i].find_element(By.CSS_SELECTOR, "span.media_journalistcard_summary_subscribe > span.media_journalistcard_summary_subscribe_inner._SUBSCRIBE_COUNT_TEXT_WRAP> em").text
            subscriber_number = re.sub(r'[^0-9]', '', subscriber_number)
            
            cheer_number = elements[i].find_element(By.CSS_SELECTOR, "span.media_journalistcard_summary_subscribe > span.media_journalistcard_summary_subscribe_inner._LIKE_HIDE").text
            cheer_number = re.sub(r'[^0-9]', '', cheer_number)
            
            names.append(name)
            subscriber_numbers.append(subscriber_number)
            cheer_numbers.append(cheer_number)
            time.sleep(2)
            
    except:
        pass

    rankingnews_journalist[url] = {'names': names, 
                                   'subscriber_numbers': subscriber_numbers, 
                                   'cheer_numbers': cheer_numbers}
```

```python
rankingnews_journalist
```

```python
rankingnews_journalist = pd.DataFrame(rankingnews_journalist).T.reset_index()
rankingnews_journalist.columns = ['rankingnews_link', 'names', 'subscriber_numbers', 'cheer_numbers']
rankingnews_journalist
```

```python
rankingnews_journalist.to_csv("rankingnews_journalist.csv", index=False)
```
